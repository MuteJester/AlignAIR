Align IG/TCR reads into an AIRR rearrangement table with the stable Python API.

When a pretrained bundle is published you will simply do `load_model("alignair/human-igh-ogrdb")`. Until then, `alignair demo` makes a tiny (not-accurate) model so the workflow is runnable end to end. **Run from the repo root.**

## Load + predict via the public API
`load_model` / `predict` / `read_sequences` are the whole inference surface.

```python
from alignair import load_model, predict, read_sequences
model = load_model("demo_out/bundle", device="cpu")
ids, reads, _ = read_sequences("demo_out/demo_reads.fasta")
result = predict(model, reads)
result.to_airr("inference_out.tsv", ids)
pd.read_csv("inference_out.tsv", sep="\t")[
    ["sequence_id", "v_call", "d_call", "j_call", "productive", "junction"]].head()
```

> The calls above are **not accurate** — this is a 1-step demo model. Train a real one (`alignair train ... --preset desktop`) or use a published bundle for real results.

## Dynamic genotype
Align against a donor reference (an allele subset and/or **novel** alleles) — no retraining.

```python
res2 = predict(model, reads, genotype="examples/novel_allele/donor_with_novel.yaml")
print(len(res2), "reads aligned against the donor genotype")
```
