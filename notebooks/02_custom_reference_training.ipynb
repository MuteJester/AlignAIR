# AlignAIR — train on your own reference (smoke)

Train a model from germline FASTA when no built-in reference fits (a non-model species, an in-house reference). This is a **smoke run** (tiny + few steps) to prove the pipeline; scale up with `--preset desktop`/`standard`. **Run from the repo root.**

```python
import sys
```

## Discover built-in references and valid chain types

```python
!{sys.executable} -m alignair.cli reference list | head -12
```

## Plan first (no training): size, time, memory, expected accuracy, warnings

```python
!{sys.executable} -m alignair.cli train --plan --device cpu --preset smoke --allow-curatable \
    --v-fasta examples/custom_reference/v.fasta \
    --j-fasta examples/custom_reference/j.fasta \
    --d-fasta examples/custom_reference/d.fasta \
    --chain-type BCR_HEAVY -o cr_model
```

## Train the smoke model

```python
!{sys.executable} -m alignair.cli train --device cpu --preset smoke --steps 8 \
    --no-calibrate --allow-curatable \
    --v-fasta examples/custom_reference/v.fasta \
    --j-fasta examples/custom_reference/j.fasta \
    --d-fasta examples/custom_reference/d.fasta \
    --chain-type BCR_HEAVY -o cr_model
```

## Predict with your custom-reference bundle

```python
from alignair import load_model, predict, read_sequences
m = load_model("cr_model/bundle", device="cpu")
ids, reads, _ = read_sequences("examples/reads.fasta")
print(len(predict(m, reads)), "reads aligned with the custom-reference model")
```

The bundle embeds your reference, so `alignair predict` needs no extra flags. For a usable model, re-run with `--preset desktop` (drop `--steps`).
